```python
import numpy as np
import pandas as pd
from scipy import stats
df=pd.read_csv("health_dataset.csv")
df.head()
```

```python
# Confidence Interval

mean_age = df['age'].mean()
std_age = df['age'].std()
n = len(df)

ci = stats.t.interval(0.95, df=n-1, loc=mean_age, scale=std_age/np.sqrt(n))
print("Confidence Interval (Age):", ci)
```

```python
# T-Test (Smoker vs Non-Smoker BMI)

smokers = df[df['smoking_status'] == 'Smoker']['bmi']
non_smokers = df[df['smoking_status'] == 'Non-Smoker']['bmi']

t_stat, p_val = stats.ttest_ind(smokers, non_smokers)

print("T-stat:", t_stat)
print("P-value:", p_val)

print("The p-value (0.9084) is significantly greater than the significance level of 0.05. Therefore, we fail to reject the null hypothesis. This indicates that there is no statistically significant difference in BMI between smokers and non-smokers.")
```

```python
# Chi-Square Test

contingency = pd.crosstab(df['smoking_status'], df['diabetes'])

chi2, p, dof, expected = stats.chi2_contingency(contingency)

print("Chi2:", chi2)
print("P-value:", p)

print("The p-value (0.0643) is slightly greater than the significance level of 0.05. Therefore, we fail to reject the null hypothesis. This suggests that there is no statistically significant association between smoking and diabetes.")
```

```python
# ANOVA Test (Age Group vs Glucose)

groups = [group['glucose_level'].values for name, group in df.groupby('age_group')]

f_stat, p_val = stats.f_oneway(*groups)

print("F-stat:", f_stat)
print("P-value:", p_val)

print("The p-value (7.33 × 10⁻¹⁴) is extremely small and much less than the significance level of 0.05. Therefore, we reject the null hypothesis. This indicates that there is a statistically significant difference in glucose levels across different age groups.")
```

```python
# Covariance & Correlation

cov = np.cov(df['age'], df['bmi'])[0][1]
corr = np.corrcoef(df['age'], df['bmi'])[0][1]

print("Covariance:", cov)
print("Correlation:", corr)

print("The positive covariance value (19.69) indicates that age and BMI tend to increase together. The correlation coefficient (0.35) suggests a moderate positive relationship between age and BMI.")
```
